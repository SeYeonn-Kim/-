```python
station_name = []
for name in crime_anal_police['관서명']:
    station_name.append('서울'+str(name[:-1])+'경찰서')
    #[관서명]에 있는 이름들을 가지고 for문 돌아서 name에
    #append()안에 있는 걸 station_name에 넣는다
    #str(name[:-1])은 name에 있던 이름의 처음부터 -1위치의 앞까지, 즉 마지막 글자를 빼고 데려옴
station_name
```

```python
station_name.remove('서울종암경찰서')
station_name
```

```python
station_address = []
station_lat = []
station_lng = []

for name in station_name:
    tmp = gmaps.geocode(name, language='ko')
    station_address.append(tmp[0].get("formatted_address"))
    
    tmp_loc = tmp[0].get("geometry")

    station_lat.append(tmp_loc['location']['lat'])
    station_lng.append(tmp_loc['location']['lng'])
    
    print(name + '-->' + tmp[0].get("formatted_address"))
```

```python
df = pd.read_excel("02. sales-funnel.xlsx")
df.head()
```

```python
pd.pivot_table(df , index = ["Name"])
#Name 항목으로만 정렬
#중복된 Name의 항목들은 하나로 합쳐지고 value들은 평균값을 가짐
```

```python
pd.pivot_table(df, index=["Name","Manager","Rep"])
#인덱스 여러개 지정 가능
```

```python
pd.pivot_table(df,index=["Manager","Rep"])
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"])
#특정 밸류만 보기 가능
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],aggfunc=np.sum)
#중복되는 value들을 평균값이 아닌 합계로 보기(sum)
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],aggfunc=[np.mean,len])
#중복되는 value들을 평균으로 보기(mean)
#len은 구성요소의 개수를 나타내는 함수
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],
               columns=["Product"],aggfunc=[np.sum])
#product로 컬럼을 지정하여 보여주기
```

```python
pd.pivot_table(df,index=["Manager","Rep"],values=["Price"],
               columns=["Product"],aggfunc=[np.sum],fill_value=0)
#Nan으로 표시된 부분을 0으로 채우기
```

```python
pd.pivot_table(df,index=["Manager","Rep","Product"],
               values=["Price","Quantity"],aggfunc=[np.sum],fill_value=0)
#value를 두 개 이상으로 잡기
```

```python
pd.pivot_table(df,index=["Manager","Rep","Product"],
               values=["Price","Quantity"],
               aggfunc=[np.sum,np.mean],fill_value=0,margins=True)
#manager, rep, product로 인덱스 지정
#price, quantity로 value 지정
#sum, mean 두가지로 value 보여주기
#Nan값은 0으로
#margins를 true로 두면 부분합을 포함 할 수 있음
```
